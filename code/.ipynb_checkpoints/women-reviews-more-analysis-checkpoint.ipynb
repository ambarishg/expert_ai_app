```python
import numpy as np
import pandas as pd
import re
import nltk
import spacy
import string
from time import time

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
pd.options.mode.chained_assignment = None
```

```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    text = str(text)
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
```

```python
FILENAME = "intimates_sentiment_traits.csv"
COLNAME ="Review Text"
NEWCOLNAME = COLNAME + "_"
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    
    text = str(text)
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    
    text = str(text)
    return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    text = str(text)
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
intimates[COLNAME + "_"] = intimates[COLNAME].apply(lambda text: remove_punctuation(text))
```

```python
intimates[COLNAME + "_"] = intimates[COLNAME + "_"].apply(lambda text: remove_stopwords(text))
```

```python
tf_feature_names
```

```python
def top_tfidf_feats(row, features, top_n=25):
    ''' Get top n tfidf values in row and return them with their corresponding feature names.'''
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.columns = ['feature', 'tfidf']
    return df
```

```python
def top_feats_in_doc(Xtr, features, row_id, top_n=25):
    row = np.squeeze(Xtr[row_id].toarray())
    return top_tfidf_feats(row, features, top_n)
```
